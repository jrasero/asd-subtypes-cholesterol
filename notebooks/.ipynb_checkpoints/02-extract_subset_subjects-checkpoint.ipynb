In this notebook, ASD and TD subjects IDs are explicitly extracted. Subjects from KUL data are discarded since it only contains ASD subjects and we want to have info from both groups per centre

```python
#IMPORTS
import numpy as np
import pandas as pd
from os.path import join as opj
import os
import pickle
```

```python
data_dir = "/home/javier/Documents/amaia-abide_sin_kul/data/"
```

```python
demo_data = pd.read_csv(opj(data_dir, "abide_I_II_demo_clean.csv"))
struct_dat = pd.read_csv(opj(data_dir, "structural.csv"))
subject_list = struct_dat.sub_id.values
```

```python
#Sites and labes
y = demo_data.loc[np.concatenate([np.where(demo_data.SUB_ID == sub)[0] for sub in subject_list]), "DX_GROUP"].values - 1
sites = demo_data.loc[np.concatenate([np.where(demo_data.SUB_ID == sub)[0] for sub in subject_list]), "SITE_ID"].values
```

```python
# here y==0 examples are ASD. It will be changed to 1 when doing the machine learning analysis
asd_subjs, tc_subjs =  subject_list[(y==0) & (sites != "KUL")], subject_list[(y==1) & (sites != "KUL")]
```

```python
print("there are then %d ASD subjects and %d TD subjects for the final analysis" % (len(asd_subjs), len(tc_subjs)))
```

```python
subjects_subgroups = { 'asd_subjs': asd_subjs, 
                      'tc_subjs': tc_subjs}
```

```python
# Save this data 
f = open(opj(data_dir, "subjects_subgroups.pkl"), "wb")
pickle.dump(subjects_subgroups, f)
f.close()
```

```python
f = open(opj(data_dir, "subjects_subgroups.pkl"), "rb")
file_loaded = pickle.load(f, encoding='utf-8')
f.close()
```
