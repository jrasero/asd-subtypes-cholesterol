```python
effs = np.zeros(shape = sparsity_matrix.shape)
trans  = np.zeros(shape = sparsity_matrix.shape)
assor  = np.zeros(shape = sparsity_matrix.shape)
lam = np.zeros(shape = sparsity_matrix.shape)
gam = np.zeros(shape = sparsity_matrix.shape)
smw = np.zeros(shape = sparsity_matrix.shape)

np.random.seed(RANDOM_STATE)
for ii in range(Y_combat.shape[0]):    
    for jj in range(sparsity_matrix.shape[1]):
        
        bin_mat = squareform((Y_combat[ii,:] > sparsity_matrix[ii, jj]).astype(int))
        bin_mat = bin_mat.astype(float)

        # simulate random network by rewiring each edge 5 times
        rand_mat = bct.randmio_und(bin_mat, itr=5)[0]

        effs[ii, jj] = bct.efficiency_bin(bin_mat)
        assor[ii, jj] = bct.assortativity_bin(bin_mat)
        trans[ii, jj] = bct.transitivity_bu(bin_mat)

        # average clustering coefficient
        c = bct.clustering_coef_bu(bin_mat).mean()
        c_r = bct.clustering_coef_bu(rand_mat).mean()

        gam[ii, jj] = c/c_r
        # average short path
        l = bct.charpath(bct.distance_bin(bin_mat),
                         include_diagonal=True, 
                         include_infinite=False)[0]
        l_r = bct.charpath(bct.distance_bin(rand_mat),
                           include_diagonal=True, 
                           include_infinite=False)[0]
        lam[ii, jj] = l/l_r

        smw[ii, jj] = gam[ii,jj]/lam[ii, jj]
    if ii % 100 == 0:
        print("node %s finished" % str(ii))
```

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.tools.tools import add_constant
```

```python
cond = (demo_net_df_thr.clus_id == 0) | (demo_net_df_thr.clus_id == 3)
cont_cols = ['AGE_AT_SCAN', 'FIQ', 'VIQ', 'PIQ', 'fc']
# the formula
temp_df = demo_net_df_thr.loc[cond,:]
y_label = 'eff'
XX = np.column_stack((pd.get_dummies(temp_df.clus_id, drop_first=True).values, 
                      pd.get_dummies(temp_df.SEX, drop_first=True).values,
                      temp_df.loc[:, cont_cols].values))
XX = add_constant(XX)
yy = temp_df.loc[:, y_label]

#formula = '%s ~ C(clus_id) + C(SEX) + AGE_AT_SCAN + FIQ + VIQ + PIQ + fc' % yy
# fit the regression model subject to the formula
reg_fit = OLS(endog=yy, exog=XX, missing='drop').fit()
```

```python
cat_cols = ['clus_id', 'SEX'] 
cont_cols = ['AGE_AT_SCAN', 'FIQ', 'VIQ', 'PIQ', 'fc']

pvals_dict = {}
beta_params = {}
for y_label in ['ass', 'trans', 'eff','lam', 'gam', 'smw']:
    for ii in range(4):
        for jj in range(ii+1, 4):
            # the condition
            cond = (demo_net_df_thr.clus_id == ii) | (demo_net_df_thr.clus_id == jj)
            # the formula
            temp_df = demo_net_df_thr.loc[cond,:]
            
            XX = np.column_stack((pd.get_dummies(temp_df.clus_id, drop_first=True).values, 
                                  pd.get_dummies(temp_df.SEX, drop_first=True).values,
                                  temp_df.loc[:, cont_cols].values))
            XX = add_constant(XX)
            yy = temp_df.loc[:, y_label]
            
            #formula = '%s ~ C(clus_id) + C(SEX) + AGE_AT_SCAN + FIQ + VIQ + PIQ + fc' % yy
            # fit the regression model subject to the formula
            reg_fit = OLS(endog=yy, exog=XX, missing='drop').fit()
            # the name of the dictionary entry
            name_case =  y_label + "_" + str(ii) + "_" + str(jj)
            # save the p-value
            pvals_dict[name_case] = reg_fit.pvalues[1]
            beta_params[name_case] = reg_fit.params[1]
```

```python
from statsmodels.stats.multitest import multipletests
pv_bonfs = multipletests(list(pvals_dict.values()), method='holm')[1]
pvals_dict_bonf = {}
for key, pv in zip(list(pvals_dict.keys()), pv_bonfs):
    pvals_dict_bonf[key] = pv
```
