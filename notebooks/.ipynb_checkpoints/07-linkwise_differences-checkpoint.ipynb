```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from statsmodels.formula import api
import ptitprince
from statsmodels.stats.multitest import multipletests
```

```python
demo_w_clusters = pd.read_excel('../data/pheno_w_clusters.xlsx')
Y_combat = np.load('../data/data_after_combat.npz')['Y_combat']
```

```python
demo_w_clusters.clus_id.value_counts()
```

```python
# Variables that we are going to control for
covariates = ['AGE_AT_SCAN', 'SEX', 'FIQ', 'VIQ', 'PIQ']
```

Define global connectivity as the average absolute connectivity and perform a linear regression to see differences among groups

```python
conn_pos = [Y_combat[ii,:][Y_combat[ii,:]>0] for ii in range(Y_combat.shape[0])]
```

```python
data_global_df_mean =  pd.concat([pd.DataFrame({'conn': [links.mean() for links in conn_pos]}),
                                  demo_w_clusters.loc[:, ['SEX', 'AGE_AT_SCAN', 'FIQ', 'VIQ', 'PIQ', 'clus_id']]]
                                 , axis=1)
```

```python
#This is using the median instead of the mean. It gives similar results
data_global_df_median =  pd.concat([pd.DataFrame({'conn':[np.median(links) for links in conn_pos]}),
                             demo_w_clusters.loc[:, ['SEX', 'AGE_AT_SCAN', 'FIQ', 'VIQ', 'PIQ', 'clus_id']]]
                             , axis=1)
```

```python
#This is using the trimmed mean. It gives similar results
from scipy.stats import trim_mean
data_global_df_trim =  pd.concat([pd.DataFrame({'conn': [trim_mean(links, proportiontocut=0.1) \
                                                         for links in conn_pos]}),
                                  demo_w_clusters.loc[:, ['SEX', 'AGE_AT_SCAN', 'FIQ', 'VIQ', 'PIQ', 'clus_id']]], 
                                 axis=1)
```

```python
reg_model = api.ols('conn ~ C(clus_id) + C(SEX) + AGE_AT_SCAN + FIQ + VIQ + PIQ', 
                    data = data_global_df_mean[data_global_df_mean.clus_id<4])
print("betas = ", np.round(reg_model.fit().params.values, 3))
print("pvalues = ", np.round(reg_model.fit().pvalues.values,3))
```
